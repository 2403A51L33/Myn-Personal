```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
```

```python
X = np.random.rand(100, 1)
```

```python
y = 3 * X + np.random.randn(100, 1)
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
print("Coefficient (slope): ", model.coef_)
print("Intercept: ", model.intercept_)
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, y, label="Data")
plt.plot(X, y_pred, color="red", label="Regression line")
plt.legend()
plt.show()
```

```python
#1. Increase or decrease noise in the data
y_noisy = 3 * X + np.random.randn(100, 1) * 2
model.fit(X, y_noisy)
y_pred_noisy = model.predict(X)
plt.scatter(X, y_noisy, label="Noisy data")
plt.plot(X, y_pred_noisy, color="red", label="Regression line (noisy)")
plt.legend()
plt.show()
```

```python
#2. Change the slope of the original function
X_new = np.random.rand(100, 1)
y_new = 6 * X_new + np.random.randn(100, 1)  # Change slope
model.fit(X_new, y_new)
y_pred_new = model.predict(X_new)
plt.scatter(X_new, y_new, label="New data")
plt.plot(X_new, y_pred_new, color="red", label="Regression line (new)")
plt.legend()
plt.show()
```

```python
#3. Use a different range for X values
X_range = np.random.rand(100, 1) * 10  # Change range
y_range = 3 * X_range + np.random.randn(100, 1)
model.fit(X_range, y_range)
y_pred_range = model.predict(X_range)
plt.scatter(X_range, y_range, label="Data (new range)")
plt.plot(X_range, y_pred_range, color="red", label="Regression line (new range)")
plt.legend()
plt.show()
```

```python
#4. Use fewer or more data points
X_few = np.random.rand(20, 1)  # Fewer data points
y_few = 3 * X_few + np.random.randn(20, 1)
model.fit(X_few, y_few)
y_pred_few = model.predict(X_few)
plt.scatter(X_few, y_few, label="Fewer data points")
plt.plot(X_few, y_pred_few, color="red", label="Regression line (fewer)")
plt.legend()
plt.show()
```

```python
#5. Non-linear relationship
X_nonlinear = np.random.rand(100, 1)
y_nonlinear = X_nonlinear ** 2 + np.random.randn(100, 1)  # Non-linear relationship
model.fit(X_nonlinear, y_nonlinear)
y_pred_nonlinear = model.predict(X_nonlinear)
plt.scatter(X_nonlinear, y_nonlinear, label="Non-linear data")
plt.plot(X_nonlinear, y_pred_nonlinear, color="red", label="Regression line (non-linear)")
plt.legend()
plt.show()
```
